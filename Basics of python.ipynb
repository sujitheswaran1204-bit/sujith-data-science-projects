### **Exercise 1: Prime Numbers**

```python

# Exercise 1: Prime Number Checker

def is_prime(num):
    """Check whether a given number is prime or not."""
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Input from user
n = int(input("Enter a number to check if it is prime: "))

# Output
if is_prime(n):
    print(f"{n} is a Prime Number.")
else:
    print(f"{n} is Not a Prime Number.")
```

### **Exercise 2: Product of Random Numbers**

```python

# Exercise 2: Product of Random Numbers

import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask user to find the product
print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Enter your answer: "))

# Check answer
if user_answer == num1 * num2:
    print(" Correct! Well done.")
else:
    print(f" Incorrect. The correct answer is {num1 * num2}.")
```

### **Exercise 3: Squares of Even or Odd Numbers**

```python

# Exercise 3: Squares of Even or Odd Numbers
# Choice: Printing squares of EVEN numbers between 100 and 200

print("Squares of EVEN numbers between 100 and 200:\n")

for i in range(100, 201):
    if i % 2 == 0:  # Even numbers only
        print(f"{i}² = {i**2}")
```

### **Exercise 4: Hospital Billing System**

```python

# Exercise 4: Hospital Billing System

# Fixed details
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300
num_days = 4
num_tests = 3
medicine_charges = 2400

# Calculate total components
room_total = room_charge_per_day * num_days
lab_total = lab_test_charge * num_tests

# Total before discount
total_before_discount = room_total + doctor_fee + lab_total + medicine_charges

# Apply 10% discount if bill exceeds ₹10,000
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

# Final bill after discount
final_bill = total_before_discount - discount

# Display detailed bill
print("----- HOSPITAL BILL -----")
print(f"Room Charges ({num_days} days): ₹{room_total}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges ({num_tests} tests): ₹{lab_total}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Total (Before Discount): ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
print("--------------------------")
```
