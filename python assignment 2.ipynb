## 1. Data Structures
Examples and basic operations for List, Tuple, Set, and Dictionary.

```python
# Lists
products = ['Pen', 'Notebook', 'Pencil', 'Eraser']
# Add a product
products.append('Sharpener')
# Remove a product
products.remove('Pencil')
# Iterate
for p in products:
    print(p)

# Tuples (immutable)
dimensions = (20, 30, 40)
print('Dimensions:', dimensions)

# Sets (unique values)
tags = {'stationery', 'office', 'school'}
tags.add('new')
print('Tags:', tags)

# Dictionary (key-value)
product_price = {'Pen': 10, 'Notebook': 50, 'Eraser': 5, 'Sharpener': 8}
# Update price
product_price['Pen'] = 12
# Add a new product
product_price['Marker'] = 25
print('Product prices:', product_price)
```

## 2. Functions
User-defined functions, default arguments, lambda, and recursion.

```python
# Simple function
def greet(name):
    return f"Hello, {name}!"

print(greet('student'))

# Function with default argument
def power(x, n=2):
    return x ** n

print('Power examples:', power(3), power(2, 3))

# Lambda function (anonymous)
square = lambda x: x * x
print('Square of 5:', square(5))

# Recursive function (factorial)
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print('Factorial of 5:', factorial(5))
```

## 3. NumPy
Array creation, operations, reshaping, and statistics.

```python
import numpy as np

# Create arrays
a = np.array([1, 2, 3, 4])
b = np.arange(10).reshape(2, 5)

print('Array a:', a)
print('Array b:\n', b)

# Arithmetic operations
print('a * 2:', a * 2)
print('a + a:', a + a)

# Statistics
print('Mean of a:', a.mean())
print('Median of a:', np.median(a))
print('Std of a:', a.std())

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print('Matrix product:\n', A.dot(B))
```

## 4. Pandas — Sales & Discount Dataset
We create a small Sales dataset and perform cleaning, descriptive analytics and visualizations (bar chart, histogram, boxplot).

```python
import pandas as pd

# Create a small Sales & Discounts dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=12, freq='M'),
    'Product_ID': [101,102,103,104,105,101,102,103,104,105,101,102],
    'Product_Name': ['Pen','Notebook','Pencil','Eraser','Marker','Pen','Notebook','Pencil','Eraser','Marker','Pen','Notebook'],
    'Sales_Amount': [250, 500, 120, 80, 300, 200, 450, 100, 90, 320, 260, 480],
    'Discount': [20, 50, 5, 0, 30, 10, 40, 0, 0, 25, 15, 45],
    'Quantity': [25, 50, 12, 8, 30, 20, 45, 10, 9, 32, 26, 48],
    'Category': ['Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery','Stationery']
}
df = pd.DataFrame(data)
# Feature: Revenue after discount
df['Revenue'] = df['Sales_Amount'] - df['Discount']

# Show basic info and first rows
print(df.info())
df.head()
```

### 4.1 Data Cleaning & Preprocessing

```python
# Check missing values
print('Missing values per column:\n', df.isnull().sum())

# Remove duplicates if any
df = df.drop_duplicates()

# Convert Date if not already
df['Date'] = pd.to_datetime(df['Date'])

# Reset index
df = df.reset_index(drop=True)

df.head()
```

### 4.2 Descriptive Statistics

```python
# Descriptive statistics for numerical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols].describe().T
```

### 4.3 Grouped Analysis: Total Revenue by Product

```python
# Total revenue by product
revenue_by_product = df.groupby('Product_Name')['Revenue'].sum().reset_index().sort_values(by='Revenue', ascending=False)
revenue_by_product
```

### 4.4 Visualizations
Bar chart for revenue by product, Histogram of discounts, Boxplot for Revenue distribution.

```python
import matplotlib.pyplot as plt

# Bar chart: total revenue by product
plt.figure(figsize=(6,4))
plt.bar(revenue_by_product['Product_Name'], revenue_by_product['Revenue'])
plt.title('Total Revenue by Product')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()

# Histogram: Discount distribution
plt.figure(figsize=(6,4))
plt.hist(df['Discount'], bins=8)
plt.title('Discount Distribution')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Boxplot: Revenue distribution
plt.figure(figsize=(6,4))
plt.boxplot(df['Revenue'])
plt.title('Revenue Boxplot')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```
